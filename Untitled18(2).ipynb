```python
import PyPDF2
import threading

def input_with_timeout(prompt, timeout=60):
    result = [None]

    def ask():
        result[0] = input(prompt)

    t = threading.Thread(target=ask)
    t.start()
    t.join(timeout)

    if t.is_alive():
        raise TimeoutError("No password entered within 60 seconds")

    return result[0]


with open(pdf_path, "rb") as f:
    reader = PyPDF2.PdfReader(f)

    if reader.is_encrypted:
        pwd = input_with_timeout("Enter PDF password: ", 60)

        if reader.decrypt(pwd) == 0:
            raise ValueError("Incorrect password")

        writer = PyPDF2.PdfWriter()

        for page in reader.pages:
            writer.add_page(page)

        with open(pdf_path, "wb") as out:
            writer.write(out)


with open(pdf_path, "rb") as f:
    p1 = PyPDF2.PdfReader(f).pages[0].extract_text() or ""

with open("page1_details.txt", "w", encoding="utf-8") as f:
    f.write(p1)
```

```python
import camelot
import pandas as pd

# Read all pages from the PDF
# 'pages="all"' tells camelot to process every page in the document
tables = camelot.read_pdf(pdf_path, pages='all', flavor='stream')

# Combine all extracted tables into a single DataFrame
# We skip the header row [1:] if it repeats on every page, or adjust based on structure
all_dfs = []
for t in tables:
    all_dfs.append(t.df)

final_df = pd.concat(all_dfs, ignore_index=True)

# Save the final DataFrame to a CSV file
output_csv = 'bank_statement_full.csv'
final_df.to_csv(output_csv, index=False)

print(f'Successfully extracted {len(tables)} tables and saved to {output_csv}')
display(final_df.head())
```
